```python
import pandas as pd
import matplotlib.pyplot as plt

path1 =  "datasets/3_normalize_datasets/canaldeventa.csv"
path2 =  "datasets/3_normalize_datasets/cliente.csv"
path3 =  "datasets/3_normalize_datasets/compra.csv"
path4 =  "datasets/3_normalize_datasets/gasto.csv"
path5 =  "datasets/3_normalize_datasets/localidad.csv"
path6 =  "datasets/3_normalize_datasets/proveedor.csv"
path7 =  "datasets/3_normalize_datasets/sucursal.csv"
path8 =  "datasets/3_normalize_datasets/tiposdegasto.csv"
path9 =  "datasets/3_normalize_datasets/venta.csv"


df_canaldeventa = pd.read_csv(path1, sep = ',', engine = 'python', encoding='utf8')
df_cliente = pd.read_csv(path2, sep = ',', engine = 'python', encoding='utf8')
df_compra = pd.read_csv(path3, sep = ',', engine = 'python', encoding='utf8')
df_gasto = pd.read_csv(path4, sep = ',', engine = 'python', encoding='utf8')
df_localidad = pd.read_csv(path5, sep = ',', engine = 'python', encoding='utf8')
df_proveedor = pd.read_csv(path6, sep = ',', engine = 'python', encoding='utf8')
df_sucursal = pd.read_csv(path7, sep = ',', engine = 'python', encoding='utf8')
df_tipodegasto = pd.read_csv(path8, sep = ',', engine = 'python', encoding='utf8')
df_venta = pd.read_csv(path9, sep = ',', engine = 'python', encoding='utf8')

```

```python
left1 = pd.merge(df_cliente, df_venta, on="id_cliente", how="left")
left2 = pd.merge(left1, df_sucursal, on="id_sucursal", how="left")
left3 = pd.merge(left2, df_canaldeventa, on="id_canaldeventa", how="left")

resultado = pd.DataFrame(left3)

resultado
```

```python
plt.plot()
x= resultado[['edad','precio']].groupby(['edad']).mean().sort_values(by='edad', ascending=False)
y = resultado.cantidad
#plt.plot(x, y, linestyle="--", color="g");
plt.plot(x)

plt.title("Ventas por rango de Edades")
plt.xlabel("Edades")          
plt.ylabel("Montos") 
plt.grid()
plt.show()
```
